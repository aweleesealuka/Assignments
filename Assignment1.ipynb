```python
'''Write a program using the following list and sort it based on the sum of the values of the tuples of the list:
[(1,5),(9,0),(12,3),(5,4),(13,6),(1,1)]'''

tuple_list = [(1,5),(9,0),(12,3),(5,4),(13,6),(1,1)]

#add the values of the tuples and sort the created list
tuple_sums = sorted([(i[0] + i[1]) for i in tuple_list])

tuple_sums
```

```python
'''Given a two Python list. Iterate both lists simultaneously such that list1 should display item in original order and list2 in reverse order:
list1 = [10, 20, 30, 40], list2 = [100, 200, 300, 400]'''

list1 = [10, 20, 30, 40] 
list2 = [100, 200, 300, 400]

#create a list of tuples of the required result using list comprehension
list3 = [(i,j) for i,j in zip(list1,list2[::-1])]
list3

#print list1 and reversed list2 simultaneously on different lines
for tup in list3:
    print (tup[0],tup[1])
```

```python
'''Given number 1 through 10, write a program to display "Even" for even numbers and "Odd" for odd numbers.'''

results = ['Even' if num%2 == 0  else 'Odd' for num in range(1,11) ]

results
```

```python
'''Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings:
['abc', 'xyz', 'aba', '1221', 'bhgsskknb', 'aa']'''

str_list = ['abc', 'xyz', 'aba', '1221', 'bhgsskknb', 'aa']

#create a list of strings in str_list that satisfy the conditions
new_list =[i for i in str_list if len(i) >= 2 and (i[0] == i[-1])]

print(len(new_list))

```

```python
'''Given a list of 1 through 10, write a program to display the square root of the even numbers'''

#create a list of the squareroot of even numbers from 1 through 10 using list comprehension
result =[num**0.5 for num in range(1,11) if num%2 == 0]

result
```
